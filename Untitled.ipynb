```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("heart.csv")
print(data.head())
```

```python
print(data.info())
```

```python
print(data.describe())
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data['age'],kde=True, bins=20,color="blue")
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x='output', y='age', data=data, hue='output', palette='Set2', dodge=False)
plt.title('Age vs. Heart Disease Output')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.legend([], [], frameon=False)
```

```python
sns.boxplot(x='output', y='chol', data=data, hue='output', palette='Set2', dodge=False)
plt.title('Cholesterol Levels vs. Heart Disease Output')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.legend([], [], frameon=False)
```

```python
corr_matr= data.corr()

plt.figure(figsize=(8,5))
sns.heatmap(corr_matr,annot=True,cmap="coolwarm",fmt=".2f",linewidths="0.5")
plt.title('Correlation Matrix Heatmap')
plt.show()
```
