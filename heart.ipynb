# the head of data

```python
data=pd.read_csv("heart.csv")
print(data.head())
```

# data with their parametrs

```python
print(data.info())
```

# the describe of data

```python
print(data.describe())
```

# Normalization

```python
data["age"]=data["age"].astype(np.uint8)
data["sex"]=data["sex"].astype(np.uint8)
data["cp"]=data["cp"].astype(np.uint8)
data["trtbps"]=data["trtbps"].astype(np.uint8)
data["oldpeak"]=data["oldpeak"].astype(np.uint8)
data["slp"]=data["slp"].astype(np.uint8)
data["caa"]=data["caa"].astype(np.uint8)
data["thall"]=data["thall"].astype(np.uint8)
data["output"]=data["output"].astype(np.uint8)

print(data.info())
```

#### after this level the size of data has become significantly smaller that caused need less space and memory

## save the results afte normalization in a cvs file

```python
data.to_csv("after_normalization.cvs")
```
